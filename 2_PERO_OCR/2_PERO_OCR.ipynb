```python
# import de la fonction process
from pipeline import process

# import de la librairie path
from pathlib import Path
```

```python
VERBOSE = True
FORCE = True
DETAILS = False
# True > on écrit chaque objet "ligne" dans un fichier différent

# Définition d'une fonction qui applique PERO a une page
def process_one_path(page: Path):
    if not page.is_file():
        return
    if page.match("*.out*"):
        return
    print(f"{10*'='} processing {page}")
    process(page, verbose=VERBOSE, force=FORCE, details=DETAILS)
```

```python
# Application de la fonction sur une page
# avec la verbose donc avec tous les détails sur cette page
process_one_path(Path("/home/aaron/Documents/M2/memoire_M2/0_Rendu_M2/output003_col2.jpg"))
```

```python
# Application de la fonction sur plusieurs pages, en enlevant la verbose
VERBOSE = False
FORCE = False
DETAILS = False

for page in Path('/home/aaron/Documents/M2/memoire_M2/annexe_pngs/crops/').glob("*.jpg"):
    process_one_path(page)
```
